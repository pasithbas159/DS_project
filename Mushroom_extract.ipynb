## ngrams

```python
from nltk import ngrams
```

```python
n = 3
n_grams = []
word_gram = []
for i in range(len(text)):
  n_grams = ngrams(text[i], n)
  for grams in n_grams:
    word_gram.append(grams)
    # print(grams)
```

## Count entire words
- ตัดคำทั้งหมดจากงานวิจัยทุกบทความ เพื่อหาความถี่เพื่อใช้อธิบาย trend หรือหัวข้อที่สนใจ

```python
text = []
for i in range(len(sentences)):
  text.append(word_tokenize(sentences[i], engine='attacut'))
  stop_words = list(thai_stopwords()) #stopwords
  stop_words.extend([' ', '(', ')', '[', ']', '.', ':'])
  text[i] = [w for w in text[i] if not w in stop_words] 
print(text[1])
```

```python
text_full = []
for i in range(len(text)):
  for j in range(len(text[i])):
    text_full.append(text[i][j])
```

```python
text_dct = {}
for i in range(len(text_full)):
  text_dct[text_full[i]] = text_full.count(text_full[i])

text_dct = {k: v for k, v in sorted(text_dct.items(), key=lambda item: item[1], reverse=True)}
text_dct
```
