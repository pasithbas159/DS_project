```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
data_file='https://github.com/prasertcbs/basic-dataset/raw/master/mnist_28x28_5000rows.csv.zip'
print(f'data_file: {data_file}')
df=pd.read_csv(data_file)
df.head()
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, :784]
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=7)
```

```python
X_train.head()
```

## PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_t = scaler.transform(X_train)
X_test_t = scaler.transform(X_test)
```

```python
X_train_t[:1]
```

```python
pca = PCA(n_components=X_train.shape[1])
```

```python
X_new = pca.fit_transform(X)
print(f'explained_variance (n_components={pca.n_components}) = {pca.explained_variance_}')
print(f'explained_variance_ratio (n_components={pca.n_components}) = {pca.explained_variance_ratio_}')
print(f'sum explained_variance_ratio = {np.sum(pca.explained_variance_ratio_)}')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, pca.n_components+1), pca.explained_variance_)
```

```python
# cvar = cumulative proportion of var explained
cvar=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=3)*100)
cvar #cumulative sum of variance explained with [n] features
```

```python
pca = PCA(.9)
pca.fit(X_train)
```

```python
pca.n_components_ #85 PCA components
```

```python
X_train_t = pca.transform(X_train)
X_test_t = pca.transform(X_test)
```

```python
X_train_t[1]
```

## model selection

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import confusion_matrix, classification_report
```

```python
algo = [
    [LogisticRegression(solver='lbfgs'), 'LogisticRegression'], 
    [SVC(), 'Support Vector Machine'], 
    [RandomForestClassifier(n_estimators=400), 'RandomForest'], 
    [AdaBoostClassifier(), 'AdaBoostClassifier']
]
model_score = []
for i in algo:
    model = i[0]
    model.fit(X_train_t, y_train)
    pred = model.predict(X_test_t)
    score = model.score(X_train_t, y_train)
    model_score.append([score, i[1]])
    print(f'{i[1]} score: {score}')
    print(confusion_matrix(y_test, pred))
    print(classification_report(y_test, pred))
    print('-'*100)
print(model_score)
```

## SVM is the best

```python
for n in [.8, .85, .9, .95]:
    scaler = StandardScaler() # Fit on training set only.
    scaler.fit(X_train) # Apply transform to both the training set and the test set.
    X_train_t = scaler.transform(X_train)
    X_test_t = scaler.transform(X_test)
    pca = PCA(n)
    pca.fit(X_train_t)
    X_train_t = pca.transform(X_train_t)
    X_test_t = pca.transform(X_test_t)
    model = SVC()
#     model = LogisticRegression(multi_class="auto", solver="lbfgs", max_iter=1000)
#     model = SVC(gamma='auto')
#     model = LinearSVC(max_iter=10000)
    model.fit(X_train_t, y_train)
    print(f'pca = {n:.2f}({pca.n_components_:3} components), accuracy (train) = {model.score(X_train_t, y_train):.4f}, accuracy (test) = {model.score(X_test_t, y_test):.4f}')
```

## RandomForest 2nd

```python
for n in [.8, .85, .9, .95]:
    scaler = StandardScaler() # Fit on training set only.
    scaler.fit(X_train) # Apply transform to both the training set and the test set.
    X_train_t = scaler.transform(X_train)
    X_test_t = scaler.transform(X_test)
    pca = PCA(n)
    pca.fit(X_train_t)
    X_train_t = pca.transform(X_train_t)
    X_test_t = pca.transform(X_test_t)
    model = RandomForestClassifier(n_estimators=400)
#     model = LogisticRegression(multi_class="auto", solver="lbfgs", max_iter=1000)
#     model = SVC(gamma='auto')
#     model = LinearSVC(max_iter=10000)
    model.fit(X_train_t, y_train)
    print(f'pca = {n:.2f}({pca.n_components_:3} components), accuracy (train) = {model.score(X_train_t, y_train):.4f}, accuracy (test) = {model.score(X_test_t, y_test):.4f}')
```
